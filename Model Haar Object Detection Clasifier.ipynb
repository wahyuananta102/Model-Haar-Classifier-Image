```python
#bahan untuk libarary yang diperlukan
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
#Deklarasi objek cascade classifier untuk deteksi bagian wajah tersebut sebagai subbahannya
face_cascade=cv2.CascadeClassifier('haarcascade_mcs_nose.xml')
```

```python
# Load image atau membaca file gambar tersebut 
img = cv2.imread('wajah tiara andini 3.bmp')
```

```python
#Konversi gambar menjadi skala abu-abu menggunakan cv2.cvtColor() dan simpan dalam variabel gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
# get the image shape ( Deteksi posisi hidung pada gambar menggunakan cascade classifier.)
rows, cols, dim = img.shape
```

**Memasukin kode untuk fungsi looping di elemen kotak tersebut**

```python
#untuk mendeteksi objek bagian hidung dalam daftar gambar grayscale
noses = face_cascade.detectMultiScale(gray, scaleFactor=1.345, minNeighbors=8, minSize=(80, 80)) 

for (x, y, w, h) in noses: #program melakukan iterasi melalui setiap kotak pembatas 
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    center = ((x + w) // 2, (y + h) // 2)
    angle = 0.8
    
    print(type(center))  
    print(type(angle)) 
```

```python
center = (img.shape[1]/2, img.shape[0]/2)
M = cv2.getRotationMatrix2D(center, angle, 0.8)
rotated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
```

```python
# apply the rotation to the image
rotated_img = cv2.warpAffine(img, M, (cols, rows)) 
```
